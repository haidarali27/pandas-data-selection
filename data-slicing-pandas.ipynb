```python
import pandas as pd

# read data from CSV
data = pd.read_csv('customers_data.csv')
data.head(4)
```

```python
# slicing using loc
data.loc[1:3]
```

```python
# slicing colomn index with loc
data.loc[1:5, "phone":"country"]
```

```python
#slicing all rows with specific colomns
data.loc[:, "city":"creditLimit"]
```

```python
#slicing using iloc
data.iloc[1:4]
```

```python
#slicing colomns with iloc
data.iloc[1:4, 2:5]
```

```python
# slicing colomns with specific name
data.iloc[1:5][["phone", "city", "country"]]
```
