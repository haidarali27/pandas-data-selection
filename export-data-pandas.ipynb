```python
import pandas as pd
```

```python
robusta_data = pd.read_csv("robusta_data_cleaned.csv")
robusta_data.head(3)
```

```python
#filter data
filtered_data_robusta = robusta_data[(robusta_data["Country.of.Origin"] == "Uganda") & (robusta_data["altitude_mean_meters"] >= 1000)]
filtered_data_robusta.head(5)
```

```python
# selected many colomns colomns
selected_cols = ["Owner", "Country.of.Origin", "Farm.Name", "Company", "Color", "Expiration", "altitude_mean_meters"]
selected_robusta_data = filtered_data_robusta[selected_cols]
selected_robusta_data.head(5)
```

```python
# rename severa colomns
rename_cols = {
    "Country.of.Origin" : "country.of.origin",
    "Farm.Name" : "farm.name"
}

rename_robusta_data = selected_robusta_data.rename(columns = rename_cols)
rename_robusta_data.head(3)
```

```python
# export data to csv
rename_robusta_data.to_csv("new_robusta_data.csv", header = False)

#read data csv
data = pd.read_csv("new_robusta_data.csv")
data.head(10)
```
